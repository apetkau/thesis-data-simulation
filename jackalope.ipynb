# Install

To install, after creating conda environment from `environment.yml` run in R:

```R
install.packages("jackalope")
```

````r
library(jackalope)

reference <- create_genome(n_chroms = 10, len_mean = 1000)
reference
````

````r
tr <- ape::rcoal(5)
tr
````

````r
ref_haplotypes <- create_haplotypes(reference, haps_phylo(tr), sub_JC69(0.1))
ref_haplotypes
````

````r
illumina(ref_haplotypes, out_prefix = "simulated", sep_files=TRUE, n_reads = 500000,
         compress=TRUE, comp_method="gzip",
         paired = TRUE, read_length = 100)
````
