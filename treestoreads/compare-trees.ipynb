```r
library(ape)
library(phytools)
library(phangorn)
library(repr)

outgroup <- "2011C-3609.fasta.wgsim"

t_orig <- read.tree(file="input/tree.tre")
t_new <- read.tree(file="phylogeny/core.aln.treefile")

t_orig <- root(t_orig, outgroup, resolve.root=TRUE)
t_new <- root(t_new, outgroup, resolve.root=TRUE)

comparePhylo(t_orig, t_new, plot=TRUE)
```

```r
treedist(t_orig, t_new, check.labels = TRUE)
```

```r
options(repr.plot.width=10, repr.plot.height=8)

plot(cophylo(t_orig, t_new, rotate=TRUE))
```
